```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/Toytoy.csv', header=None) #load the csv file --> the "history table"
df = df.reset_index(drop = True)
df.head()
```

```python
df_new = df.loc[:,1:]
items = pd.unique(df_new.values.flatten()).tolist() #generate a list of unique item ID
items_sorted = sorted(items)
items_sorted #sort the list alphabetically 
```

```python
co_matrix = np.zeros((len(items_sorted), len(items_sorted))) #create a covisitation matrix template
co_matrix = pd.DataFrame(co_matrix)
co_matrix.index = items_sorted #label the rows with the list of unique item ID
co_matrix.columns = items_sorted #label the columns ...
co_matrix
```

```python
#display all the pairs in the matrix
for i in co_matrix.index: 
  for j in co_matrix.columns:
    co_matrix[i][j] = (i+j)

co_matrix
```

```python
df_combine = []

for i in np.array(df_new): #create a list of lists that contains all the pairs in the "history table"
    combined_strings = []
    for j in i:
        combined_strings.append(j)
    df_combine.append(combined_strings)

df_combine
```

```python
bigrams_list = []

for i in range(len(df_combine)): #find all the possible pairs from the "history table"
    current_bigrams = [(df_combine[i][j], df_combine[i][j+1]) for j in range(len(df_combine[i])-1)]
    bigrams_list.extend(current_bigrams)

print(len(bigrams_list))
print(bigrams_list)
```

```python
from collections import Counter
pairs_counter = Counter(bigrams_list) #count each unique pairs (i.e. order matters) from the "history table"
all_pairs = pd.DataFrame({'Pairs': list(pairs_counter.keys()), 'Pairs_Count': list(pairs_counter.values())})

all_pairs['Strings'] = [str(i[0])+str(i[1]) for i in all_pairs['Pairs']] #create a column of "Strings" to match the format of the pairs in the matrix
all_pairs

all_pairs_sorted = all_pairs.sort_values(by='Pairs_Count', ascending=False) # for view only
all_pairs_sorted
```

```python
for i in range(len(co_matrix)): #complete the covisitation matrix by replacing all the pairs with their corresponding values
  for j in range(len(co_matrix)):
    if len(all_pairs[all_pairs['Strings'] == str(co_matrix.iloc[i,j])]) > 0:
      co_matrix.iloc[i,j] = all_pairs[all_pairs.iloc[:, 2] == str(co_matrix.iloc[i,j])].iloc[0, 1]
    else:
      co_matrix.iloc[i,j] = 0

co_matrix
#this matrix is created based on line 55. If there is anything changes in the matrix, be sure to run line 55 again.
```
